```python
guess_list = ['Almond', 'Banana', 'Blewits', 'Brazil', 'Carrot', 'Casaba', 'Cashew', 'Celery', 'Cherry', 'Chives', 'Chocho', 'Citron', 'Citrus', 'Cob nut', 'Cooker', 'Daikon', 'Damson', 'Endive', 'Fennel', 'Garlic', 'Girkin', 'Greens', 'Kiwano', 'Lentil', 'Lichee', 'Lychee', 'Marrow', 'Medlar', 'Nettle', 'Orange', 'Oyster', 'Papaya', 'Pawpaw', 'Peanut', 'Pepper', 'Pignut', 'Pippin', 'Potato', 'Quince', 'Radish', 'Raisin', 'Rocket', 'Russet', 'Squash', 'Tomato', 'Turnip', 'Wakame', 'Walnut']
```

```python
print(game_name)
counter = 0

game_again = False

import random
while not game_again:
    game_over = False
    word = random.choice(guess_list).lower()
    word_list = list(word.lower())
    user_guess_list = ['_ ' for i in range(len(word))]
    
    while not game_over:
        letter = input('Enter the letter you guess:\n').lower().strip()
        if letter in word:
            for i in range(len(user_guess_list)):
                if word_list[i] == letter:
                    user_guess_list[i] = letter
            print(''.join(user_guess_list))
            if ''.join(user_guess_list) == word:
                print('Congrats! You won.')
                game_over = True
        else:
            print(hung_status[counter])
            if counter == 7:
                print('Sorry! You lost.')
                game_over = True
            else:
                counter += 1
                continue
    c = input('Do you want to continue with this game? (y/n):\n').strip().lower()
    if c == 'y':
        continue
    else:
        game_again = True
```
