```python
import random

board = [' ' for i in range(9)]

def print_board():
    print('|' + board[0] + '|' + board[1] + '|' + board[2] + '|')
    print('|' + board[3] + '|' + board[4] + '|' + board[5] + '|')
    print('|' + board[6] + '|' + board[7] + '|' + board[8] + '|')

def player_select():
    player_list = ['Player1', 'Player2']
    player = random.choice(player_list)
    print(f'You can start play {player}')
    global d
    if player == 'Player1':
        d = 'Player2'
    else:
        d = 'Player1'
    return player

def play(p):
    print(f'Its your turn- {p}')
    n = int(input('Enter the column number from 1 to 9 in the board you select:\n'))
    if p == 'Player1':
        board[n-1] = 'X'
    elif p == 'Player2':
        board[n-1] = 'O'
    print_board()

def P1_win(b1):
    if (b1[0] == b1[1] == b1[2] == 'X') or (b1[3] == b1[4] == b1[5] == 'X') or (b1[6] == b1[7] == b1[8] == 'X') or (b1[0] == b1[3] == b1[6] == 'X') or (b1[1] == b1[4] == b1[7] == 'X') or (b1[2] == b1[5] == b1[8] == 'X') or (b1[0] == b1[4] == b1[8] == 'X') or (b1[2] == b1[4] == b1[6] == 'X'):
        return True
    return False

def P2_win(b2):
    if (b2[0] == b2[1] == b2[2] == 'O') or (b2[3] == b2[4] == b2[5] == 'O') or (b2[6] == b2[7] == b2[8] == 'O') or (b2[0] == b2[3] == b2[6] == 'O') or (b2[1] == b2[4] == b2[7] == 'O') or (b2[2] == b2[5] == b2[8] == 'O') or (b2[0] == b2[4] == b2[8] == 'O') or (b2[2] == b2[4] == b2[6] == 'O'):
        return True
    return False

def draw(b3):
    if b3[0] != ' ' and b3[1] != ' ' and b3[2] != ' ' and b3[3] != ' ' and b3[4] != ' ' and b3[5] != ' ' and b3[6] != ' ' and b3[7] != ' ' and b3[8] != ' ':
        return True
    return False

print('Welcome to Tic Tac Toe')
print('Player1 will be X and Player2 will be O. First play will be randomly selected')
#game_over = False

print_board()
c = player_select()
while True:
    play(c)
    
    if P1_win(board):
        print('Player1 (X) wins!')
        break
    if P2_win(board):
        print('Player2 (O) wins!')
        break
    if draw(board):
        print('Its a draw.')
        break
    
    play(d)
    
    if P1_win(board):
        print('Player1 (X) wins!')
        break
    if P2_win(board):
        print('Player2 (O) wins!')
        break
    if draw(board):
        print('Its a draw.')
        break
```
