```python
import os
os.chdir(r'C:\Users\sreeraj.va\Documents\images_jpg')
```

```python
# conveting jpg images to png in bulk

for i in os.listdir('.'):
    if i.endswith('.jpg'):
        j = Image.open(i)
        fn, fext = os.path.splitext(i)
        j.save('images_png\{}.png'.format(fn))
```

```python
# changing size of an image

size_300 = (300,300)

for i in os.listdir('.'):
    if i.endswith('.jpg'):
        j = Image.open(i)
        fn, fext = os.path.splitext(i)
        
        j.thumbnail(size_300)
        j.save('size_change_300\{}.{}'.format(fn, fext))
```

```python
img2 = Image.open(r'dog.jpg')
```

```python
img2
```

```python
# rotating a picture

img2.rotate(90)
```

```python
# rotate and save

img2.rotate(270).save('dog_rotate_270.jpg')
```

```python
# to make a picture black and white

img2.convert(mode='L')
```

```python
img2.convert(mode='L').save(r'dog_bw.jpg')
```

```python
# to blur an image

img2.filter(ImageFilter.GaussianBlur(15))
```

```python
img2.filter(ImageFilter.GaussianBlur(15)).save(r'dog_blur.jpg')
```
