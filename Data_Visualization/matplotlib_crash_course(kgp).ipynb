# matlab vs matplotlib

```python
#object oriented method (for matplotlib programming)
```

```python
fig = plt.figure()    #options :  num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None, frameon=True, 
                       #FigureClass=<class 'matplotlib.figure.Figure'>, clear=False
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1]) #x & y axes values start from 0.1 to 1
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1])
axes.plot(x,y, 'ro-')
axes.set_xlabel('x data')
axes.set_ylabel('y data')
axes.set_title('Object oriented way - matplotlib-sample plot')
```

```python
dir(axes)  #things we can apply in axes
```

```python
#multiple plots together
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.3])

ax1.plot(x,y, 'ro-')
ax2.plot(x,y2, 'b*--')
ax1.set_label('x')
ax1.set_label('y')
ax1.set_title('Multiple plot in 1 fig')
#plt.axis([0,10,0,50])
```

```python
fig, ax = plt.subplots(1,2)
col = ['r', 'b']
data = [y,y2]
#limit = [[0,10,0,50],[0,10,0,500]]

for i, axes in enumerate(ax):
    axes.plot(x,data[i], col[i])
#    plt.axis(limit[i])  #1st section not taking- need to check

ax[0].set_xlim([0,10])
ax[0].set_ylim([0,50])
ax[1].set_xlim([0,10])
ax[1].set_ylim([0,500])


fig.tight_layout()
```

```python
# to change figure size
fig, ax = plt.subplots(figsize=(8,4), dpi=100)    # options :  num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None, frameon=True,
                        #FigureClass=<class 'matplotlib.figure.Figure'>, clear=False
```

```python
from matplotlib import ticker
```

```python
fig, ax = plt.subplots()
ax.plot(x,y2)
ax.set_title('Scientific notation')
```

```python
fig, ax = plt.subplots()
ax.plot(x,y2)
ax.set_title('Scientific notation')

formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1,-1))
ax.yaxis.set_major_formatter(formatter)   #formatter to format axis
```
