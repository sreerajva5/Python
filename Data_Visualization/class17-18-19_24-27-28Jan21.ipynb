```python
df = pd.DataFrame({'dt':ts, 'sales':np.random.randint(10,50,30)})
```

```python
df
```

```python
df.set_index('dt', inplace=True)
```

```python
df.head()
```

```python
df.plot()
```

```python
# to adjust the length and height of the plot

df.plot(figsize=(15,5))

# figsize is a tuple with first element as length and second elemnt represents height
```

```python
# to give a title for the plot

df.plot(figsize=(15,5), title='Day wise sales trend')


# we can pass title arguement, which is a string
```

```python
# to set labels for X-axis and Y-axis

ax = df.plot(figsize=(15,5), title='Day wise sales trend')
ax.set_xlabel('Date')
ax.set_ylabel('Sales count')
```

```python
# to change the color of the line

ax = df.plot(figsize=(15,5), title='Day wise sales trend', c='r')
ax.set_xlabel('Date')
ax.set_ylabel('Sales count')

# c=r represents color='red'

# b - blue
# k - black
# y - yellow
# g - green
# m - magenta
# w - white
# c - cyan
```

```python
# to add markers in the line graph

ax = df.plot(figsize=(15,5), title='Day wise sales trend', c='r', marker='^')
ax.set_xlabel('Date')
ax.set_ylabel('Sales count')

# commonly using marker sybol are 'o' - circle, '*' - *, '^' - triangle, 'v'  triangle down, 's'- square, 'x' -x
```

```python
# to change the size of the marker

ax = df.plot(figsize=(15,5), title='Day wise sales trend', c='r', marker='^', markersize=10)
ax.set_xlabel('Date')
ax.set_ylabel('Sales count')
```

```python
# to change the marker color

ax = df.plot(figsize=(15,5), title='Day wise sales trend', c='r', marker='^', markersize=10, markerfacecolor='b')
ax.set_xlabel('Date')
ax.set_ylabel('Sales count')
```

```python
# bar plot

df.plot(figsize=(15,5), title='Day wise sales trend', kind='bar')
```

```python
# bar plot horizontal

df.plot(figsize=(10,10), title='Day wise sales trend', kind='barh')
```

```python
# box plot

df.plot(figsize=(15,5), title='Day wise sales trend', kind='box')

# boxplot helps to identify if the data is normally distributed
```

```python
x = [1,3,2,5,10,8,12,15]

y = list(range(1,9))
```

```python
# object oriented way in matplotlib

fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])  # [start postion x-axis, start postion y-axis, width x-axis, width y-axis]
axes.plot(x,y)
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('title')
plt.show()
```
